## 1. Problem Statement

Consider the Earth and the atmosphere being two seperate bodies that absorbs and emit thermal energy from and to each other. Model the temperature of the Earth and the atmosphere over time.

## 2. Variables and Parameters

| Description | Symbol | Dimensions | Type |
| --- | ---| --- | --- |
| temperature of the Earth | $T_E$ | $\Theta$ | dependent variable |
| temperature of the atmosphere | $T_A$ | $\Theta$ | dependent variable |
| time | $t$ | T | independent variable |
| heat capacity of the Earth and atmosphere | $C$ | M L<sup>2</sup> T<sup>-2</sup> $\Theta^{-1}$ | parameter |
| albedo of the Earth | $\alpha$ | 1 | parameter |
| radius of the Earth and the atmosphere | $R$ | L | parameter |
| solar constant | $S_0$ | M T<sup>-3</sup> | parameter |
| greenhouse parameter | $\varepsilon$ | 1 | parameter |
| fraction of the energy that is emitted to the outerspace rather than earth | $\beta$ | 1 | parameter |
|Stefan Boltzmann Constant | $\sigma$| M $T^{-3}$ $\Theta^{-4}$ | constant |

## 3. Assumptions and Constraints

1. Treat the radius of the Earth and that of the Atmosphere the same.
2. Both the Earth and atmosphere emit thermal energy like black bodies.
3. The radius of the Earth is homogeneous. That is, the Earth is a perfect sphere.
4. The temperature across the Earth is homogeneous.
5. The Solar constant is a constant.
6. The albedo of the Earth is a constant.
7. The greenhouse parameter is a constant.
8. The fraction of the thermal energy emitted to the outerspace is constant.
9. The atmosphere does not absorb any shortwave thermal energy from the Sun.

## 4. Build Solutions

We have the govering equation,
$$C \frac{dT}{dt} = Q_{in} - Q_{out}$$ 

For the Earth, thermal energy comes from the Sun and the atmosphere, that is 
$$Q_{in} = Q_{Sun} + Q_{atmos}$$

By the information provided, we have $$Q_{Sun} = (1 - \alpha) \pi R^2 S_0$$

For $Q_{atmos}$, we can apply the Stefan–Boltzmann Law, and considering that a fraction of $\beta$ would be emitted to the outerspace, the part that goes to the Earth would be $$Q_{atmos} = (1-\beta) 4 \pi R^2 \sigma T_A^4$$

So, combine these two parts together,
$$Q_{in} = Q_{Sun} + Q_{atmos} = (1 - \alpha) \pi R^2 S_0 + (1-\beta) 4 \pi R^2 \sigma T_A^4$$

The thermal energy emits like black body, so by the Stefan–Boltzmann Law we have,
$$Q_{out} = 4 \pi R^2 \sigma  T_E^4$$

Altogether, 
$$C_E \frac{dT_E}{dt} = (1 - \alpha) \pi R^2 S_0 + (1-\beta) 4 \pi R^2 \sigma T_A^4 - 4 \pi R^2 \sigma  T_E^4$$ 

For the atmosphere, thermal energy comes from the Earth. We should also take into consideration that only a fraction of $\varepsilon$ woule go to the outerspace, a fraction of $(1 - \varepsilon)$ is absorbed by the atmosphere. So that gives,
$$Q_{in} = (1-\varepsilon) 4 \pi R^2 \sigma T_E^4$$

The thermal energy emits like black body so by the Stefan–Boltzmann Law,
$$Q_{out} = 4 \pi R^2 \sigma  T_A^4$$

Altogether, 
$$C_A \frac{dT_A}{dt} = (1-\varepsilon) 4 \pi R^2 \sigma T_E^4 -4 \pi R^2 \sigma  T_A^4 $$ 

So finally, we reached to a system of differentical equations,

$$
\begin{align*}
C_E \frac{dT_E}{dt} &= (1 - \alpha) \pi R^2 S_0 + (1-\beta) 4 \pi R^2 \sigma T_A^4 - 4 \pi R^2 \sigma  T_E^4 \\
C_A \frac{dT_A}{dt} &= (1-\varepsilon) 4 \pi R^2 \sigma T_E^4 -4 \pi R^2 \sigma  T_A^4 
\end{align*}
$$ 

### 4.1 Nondimentionalization

We need to scale $T_E$, $T_A$ and $t$.  

Let $T_E = [T] T_E^*$, $T_A = [T] T_A^*$ and $t = [t]t^*$. Here, I choose the same scale for $T_E$ and $T_A$ because they are both temperatures, and in the equation, their roles are very similar.

Plugging-in,

$$
\begin{align*}
C_E \frac{[T]}{[t]}\frac{dT_E^*}{dt^*} &= (1 - \alpha) \pi R^2 S_0 + (1-\beta) 4 \pi R^2 \sigma [T]^4 T_A^{*4} - 4 \pi R^2 \sigma  [T]^4 T_E^{*4} \\
C_A \frac{[T]}{[t]}\frac{dT_A^*}{dt^*} &= (1-\varepsilon) 4 \pi R^2 \sigma [T]^4 T_E^{*4} -4 \pi R^2 \sigma  [T]^4 T_A^{*4} 
\end{align*}
$$ 

Note that the highest order terms are: $T_E^{*4}$ and $T_A^{*4}$. Deviding by $4 \pi R^2 \sigma  [T]^4$ gives,

$$
\begin{align*}
\frac{C_E [T]}{[t] 4 \pi R^2 \sigma  [T]^4 }\frac{dT_E^*}{dt^*} &= \frac{(1 - \alpha) \pi R^2 S_0}{4 \pi R^2 \sigma  [T]^4} + (1-\beta) T_A^{*4} - T_E^{*4} \\
\frac{C_A [T]}{[t] 4 \pi R^2 \sigma  [T]^4 }\frac{dT_A^*}{dt^*} &= (1-\varepsilon) T_E^{*4} - T_A^{*4} 
\end{align*}
$$ 

The right-hand side of the 1st equation give us 
$$[T] = \left( \frac{(1-\alpha)S_0}{4 \sigma} \right)^{1/4} $$

Setting the left-hand side of the 1st equation to be 1 gives us,
$$[t] = \frac{C_E}{4 \pi R^2 \varepsilon} \left( \frac{(1-\alpha)S_0}{4 \sigma} \right)^{-3/4}$$

Our nondimensionalized result is,
$$
\begin{align*}
\frac{dT_E^*}{dt^*} &= 1 + (1-\beta) T_A^{*4} - T_E^{*4} \\
\frac{C_A}{C_E} \frac{dT_A^*}{dt^*} &= (1-\varepsilon) T_E^{*4} - T_A^{*4} 
\end{align*}
$$

From this result, we can see that 
* $T^*$ is a multiple of some combinations of $\alpha$, $S_0$, and $\varepsilon$, and that $t^*$ is a multiple of some combinations of $\alpha$, $S_0$, $\varepsilon$, and $C_E$.
* The parameters $\alpha$, $S_0$, and $\varepsilon$ does not matter much: they just change the scale of the model.
* The ratio $\frac{C_A}{C_E}$, $\beta$ and $\varepsilon$ changes the behavior of the model because they remains in the nondimensionalized result.

### 5.1 Numerical Solutions

#### 5.1.1 Solutions for different initial conditions

Let's find the solutions for different initial conditions.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi

print("packages imported")
```

```python
def f(u,t):
    dudt = np.zeros(2)
    dudt[0] = 1 + (1-beta)* u[1]**4 - u[0]
    dudt[1] = ce / ca * ((1-epsilon)* u[0]**4 - u[1])
    return dudt
```

```python
beta = 0.9
epsilon = 0.9
ce = 10
ca = 7

t = np.linspace(0,10,100)
ini_conds = [[0,0],[0,1],[0.5,1],[1,0.5],[1,0.1],[1,1],[1.2,1.1],[1.5,1.5]]
for u0 in ini_conds:
    U = spi.odeint(f,u0,t)
    plt.plot(t,U[:,0],'b')
    plt.plot(t,U[:,1],'g')
plt.xlabel('$t^*$')
plt.ylabel('$T^*_A, T^*_E$')
plt.title("Plot 1")
plt.grid(True)
plt.show()
```

### 5.2 Analysis

**The influence of initial conditions**  
From plot 1 and 2, it seems that $T_E^*=1.0$ and $T_A^*=0.1$ is the steady state solution. This is practical because we indeed expect some equilibrium to reach. However, plot 3 tells us that if the initial temperature goes "large", althogh just from 2 to 3, it would "blow up". That may because the scale is large, and $T^*\ge3$ will never occur in physics.  

**The influence of $\beta$**  
From plot 4, we see that the value of $\beta$ does not matter much: it only changes a bit the rate of change of the temperatures, even if we take extreme values like 0.0 and 1.0. That's maybe reasonable to simplify this model by picking a fixed value of $\beta$.

The result is some kind of reasonable. Recall that $\beta$ is the fraction of thermoenergy that goes to the outerspace and a proportion of (1 - $\beta$ ) is reabsorbed by the Earth. That energy can be see a loss or increase in the energy from the Sun. Because $S_0$ does not influence the behavior of the model, it is shown that $\beta$ does not influence much either.

**The influence of $\varepsilon$**  
From plot 5, it is shown that $\varepsilon$ do influence the equilibrium temperature of $T_A$, but not that of $T_E$. And that make sence because the physical meaning of $\varepsilon$ is the greenhouse parameter. This would in turn influence $T_E$, but for the same reason mentioned in the last section, this would not be a big difference.

Also, we exprect if $\varepsilon$ increase, $T_A^*$ decrease if you think with green house effect. And that is indeed the result(see plot 6 where the light green one is the one with larger $\varepsilon$)

It is also demonstrated that if the greenhouse effect being too wild ($\varepsilon$ too small), some "blow up" would occur (see plot 7 for case for $\varepsilon=0.3$).

**The influence of $\frac{C_E}{C_A}$**  
From plot 8, we see that the ratio $\frac{C_E}{C_A}$ does not influence much; it only affects the rate of increase of ${C_A}$. That's out of my expectation. But that's what it reflects. If I consider that this is just a scale for the second differential equation, it's somehow reasonable.
