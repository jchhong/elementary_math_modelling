#### Preliminaries
Many of the models we have discussed or simulated require input parameters, which are often measured quantities. Measured quantities are often said to have a level of uncertainty associated with them. Even if the measurement is *perfectly* done, there still exists an uncertainty with the result (e.g., perhaps the way we are measuring the quantity is inaccurate or the tools we are using to measure have uncertainty associated with them). The level of uncertainty can influence the output of any model we contruct. 

We know our measurements will have some uncertainty associated with them, and we want to determine how this uncertainty affects the output of the model.  The general strategy for determining how uncertainty influences the model is through the Monte Carlo scheme below.  

Let $x$ be the measured quantity with uncertainty $c$; i.e., $x \pm c$. 

1. Define a random variable $\hat{x} = x + \varepsilon$, where $\varepsilon$ is a normally distributed random variable with mean $0$ and standard deviation $c$. 
2. Solve the model with $\hat{x}$.
3. Repeat until the desired number of simulations is met. 
4. Visualize results.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

print("packages imported")
```

First, let's analyze the day Oct 31 (t=30).

Note: for simplicity, I use Bi to reperesent $\beta_i$ in the code blocks

According to the information in the preliminaries, we should use $\beta$s as a normally distributed random variable with the standard deviation being the $c$s.

```python
np.random.seed(111)

# first analyze the Oct.31: t=30

N = 10000
t = 30
B1 = 3
B2 = 5 + stats.norm.rvs(loc=0,scale=0.5,size=N)
B3 = 25 + stats.norm.rvs(loc=0,scale=10,size=N)
B4 = 2.35 + stats.norm.rvs(loc=0,scale=0.65,size=N)
B5 = 1.75 + stats.norm.rvs(loc=0,scale=0.85,size=N)
B6 = 3.5 + stats.norm.rvs(loc=0,scale=1.2,size=N)

Y = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

plt.hist(Y,bins = 40, density = True, alpha=0.5) 
plt.title('Distribution of Net Income')
plt.show()
```

### Task 3
3. Determine the probability density function of the Net Income on October 31 and determine the probability that you make a profit on Halloween. 

Let's use the model where only include selected uncertainties.

```python
plt.hist(Y2,bins=40,density=True,alpha=0.5)

kde_Y2 = stats.gaussian_kde(Y2)

x = np.linspace(0,8000,1000)
plt.plot(x,kde_Y2(x)), plt.grid()
plt.axvline(0, color='peru')
plt.title('Kernel Density Function (Gaussian)')
plt.show()
```

Compute the probabillity of making profit on Halloween, $\Pr(Y) > 0$

```python
p = kde_Y2.integrate_box_1d(0,np.inf)
p
```

We can see that the probability of making profit on Halloween is 0.99926.

We are also interested in the expected net income with some randomness. So let's compute the 90% confidence interval for $Y$.

```python
mu = Y2.mean()

for z in range(1500,2500,100):
    pr = kde_Y2.integrate_box_1d(mu-z,mu+z)
    print(mu-z, '\t', mu+z, '\t', pr)
```

So the 90% confidence interval is (1513, 5313)

## 5. Analysis

Let's analyse the effect for different days (different $t$ values). Let's still consider the model with selected uncertainties.

```python
np.random.seed(111)

# the Oct 1: t=0

N = 10000
t = 0
B1 = 3
B2 = 5
B3 = 25 + stats.norm.rvs(loc=0,scale=10,size=N)
B4 = 2.35
B5 = 1.75
B6 = 3.5 + stats.norm.rvs(loc=0,scale=1.2,size=N)

Y1 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

t = 10
Y11 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

t = 15
Y16 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

t = 20
Y21 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

t = 25
Y26 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

t = 30
Y31 = B1*B2*B3 + B6*np.square(t) - B4*t -B5*np.sqrt(t)

plt.hist(Y1,bins = 40, density = True, alpha=0.5, label='Oct 1') 
plt.hist(Y11,bins = 40, density = True, alpha=0.5, label='Oct 11') 
plt.hist(Y16,bins = 40, density = True, alpha=0.5, label='Oct 16') 
plt.hist(Y21,bins = 40, density = True, alpha=0.5, label='Oct 21') 
plt.hist(Y26,bins = 40, density = True, alpha=0.5, label='Oct 26')
plt.hist(Y31,bins = 40, density = True, alpha=0.5, label='Oct 31') 
plt.title('Distribution of Net Income for different days'), plt.legend()
plt.show()
```

Then let's print the expected net incomes for these days. Because these distributions are overall symmetric, let's just use the mean to approximate the expection.

```python
print('Date \t Expected Net Income', )
print('Oct 1 \t', Y1.mean())
print('Oct 11 \t', Y11.mean())
print('Oct 16 \t', Y16.mean())
print('Oct 21 \t', Y21.mean())
print('Oct 26 \t', Y26.mean())
print('Oct 31 \t', Y31.mean())
```
